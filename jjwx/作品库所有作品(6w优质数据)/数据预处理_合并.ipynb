```python
import pandas as pd
import numpy as np
```

```python
main = pd.read_csv('E:\爬虫\jjwx\作品库所有作品（数据少但是优质）\作品库.csv')
main
```

```python
author = pd.read_csv('E:\爬虫\jjwx\作品库所有作品（数据少但是优质）\作者信息.csv')
author
```

```python
main_author = pd.merge(main, author, on='a_作者', how='left')
main_author
```

```python
piece = pd.read_csv('E:\爬虫\jjwx\作品库所有作品（数据少但是优质）\文章信息.csv')
piece['piece_url'] = piece['piece_url'].apply(lambda x: x[22:].strip())
piece = piece.rename(columns={'piece_url': 'b_作品_url'})
piece
```

```python
data = pd.merge(main_author, piece, on='b_作品_url', how='left')
data
```

```python
data = data.drop(['类型_1', '类型_2', '类型_3', '类型_4', 'g_作品积分', 'b_作品', 'e_进度', 'd_风格', 'f_字数'], axis=1)
data
```

```python
data.columns = ['a_作者', 'author_url','piece_url', 'b_发表时间', 'c_作品库类型', 'a_被收藏数', 'a_最近更新作品','a_作品状态', 'a_作品字数',
                'a_最后更新时间', 'a_作者所发送红包数', 'a_小说完本数', 'a_小说连载数', 'a_小说暂停数', 'b_作品视角', 'b_作品风格', 'b_全文字数',
                'b_前进一名所需地雷数', 'b_总书评数', 'b_总共地雷数量', 'b_总点击量', 'b_文章名称', 'b_文章当前被收藏数', 'b_文章积分','b_文章类型',
                'b_文章进度', 'b_签约状态', 'b_营养液数', 'b_霸王票全站排行']
```
